# Findings
#### Salary columnn have -1, Employer Provided Salary:$150K-$160K etc


```python
df[df['Salary Estimate']=='-1']
```

```python
df.shape
```

```python
df=df[df['Salary Estimate']!='-1']
```

```python
salary= df['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
salary
```

```python
minus_kd=salary.apply(lambda x : x.replace('$','').replace('K',''))
```

```python
minus_kd
```

```python
df['hourly']=df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
```

```python
df[df['hourly']==1]
```

```python
df['Employer_provide_salary']=df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)
```

```python
minus_kd.unique()
```

```python
minus_kd=minus_kd.apply(lambda x : x.lower().replace('employer provided salary:', ''))
```

```python
minus_kd=minus_kd.apply(lambda x: x.lower().replace('per hour',''))
```

```python
minus_kd=minus_kd.apply(lambda x: x.strip())
```

```python
minus_kd.unique()
```

```python
min_salary=minus_kd.apply(lambda x: int(x.split('-')[0]))
max_salary=minus_kd.apply(lambda x: int(x.split('-')[1]))
```

```python
min_salary
```

```python
max_salary
```

```python
avg_salary=(min_salary+max_salary)/2
```

```python
df['Max Salary']=max_salary
df['Min Salary']=min_salary
df['Avg Salary']=avg_salary
```

```python
df.head()
```

### Cleaning Company name as they have number in theeir name.

```python
company_name=df['Company Name'].apply(lambda x: x.split('\n')[0])
```

## Age of the Company

```python
df['Founded'].unique()
```

```python
df['Age of Company']=df['Founded'].apply(lambda x: x if x==-1 else 2025-x)
```

```python
df['aws']=df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
```

```python
df['excel']=df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0)
```
